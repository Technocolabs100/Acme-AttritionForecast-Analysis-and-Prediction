**3. Data Encoding**

```python
df.head(3)
```

```python
df['Attrition'].unique()
```

```python
df['Attrition'].value_counts()
```

```python
# df['Attrition'] = df['Attrition'].map({'Yes':1, 'No':0})
df['Attrition'] = df['Attrition'].replace({'Yes':1, 'No':0})
```

```python
df['Over18'].unique()
```

```python
df['Over18']=df['Over18'].replace({'Y':1})
```

```python
df['OverTime'].unique()
```

```python
df['OverTime']=df['OverTime'].replace({'Yes':1,'No':0})
```

```python
df['MaritalStatus'].unique()
```

```python
df['MaritalStatus'] = df['MaritalStatus'].replace({'Single': 0, 'Married': 1, 'Divorced': 2})
```

```python
df['Gender'].unique()
```

```python
df['Gender'] = df['Gender'].replace({'Female': 0, 'Male':1})
```

```python
df['Department'].unique()
```

```python
df['Department'] = df['Department'].replace({'Sales': 0, 'Research & Development':1, 'Human Resources':2})
```

```python
df['BusinessTravel'].unique()
```

```python
df['BusinessTravel'] = df['BusinessTravel'].replace({'Travel_Rarely':0, 'Travel_Frequently':1, 'Non-Travel':2})
df.head(3)
```

**4. Data Labelling**

```python
df.info()
```

```python
one_hot_encoded = pd.get_dummies(df[['EducationField', 'JobRole']], prefix='category')
one_hot_encoded = one_hot_encoded.astype(int)
df_encoded = pd.concat([df, one_hot_encoded], axis=1)
```

```python
df_encoded.head(3)
```

```python
df_encoded.head(3)
```

```python
df_encoded.drop(['EducationField', 'JobRole'], axis=1, inplace=True)
```
